## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

NoSQL is not relational - that means it doesn't have to follow the relational format and structure. It can be stored in many different ways, and the person storing it doesn't necessarily have to know the structure beforehand. 
Neither is better; they're used for different purposes.





## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

A NoSQL database would likely be the best fit for a scenario where every document or entry has a different structure. 




## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MongoDB is a document store database. Each entry is a document, and each document stores key-value pairs. The data is stored as the value, and the key is queried against.
This is different from a relational database for the same reasons listed above - it's dynamic and unstructured. Each document can have a different structure from the other documents.






## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance defines a database's ability to deliver accurate and complete results while one or more components is unresponsive. RDDs must be fault tolerant, even though they're distributed. Applications like Hadoop and Spark are used for this purpose.






## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MapReduce refers to two different things - map AND reduce.
Map refers to running a query in a distributed system on each server simultaneously.
Reduce refers to combining the results of those distributed queries.
